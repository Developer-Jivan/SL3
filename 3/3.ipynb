```python
# Descriptive Statistics using Titanic Dataset (data.csv)

# Step 1: Import Libraries
import pandas as pd
import numpy as np

# Step 2: Load Dataset
df = pd.read_csv("data.csv")

# Step 3: Display First 5 Rows
print("First 5 Rows:")
print(df.head())

# Step 4: Check Dataset Information
print("\nDataset Info:")
print(df.info())

# Step 5: Select Required Columns
# Categorical Variable = Sex
# Numeric Variables = Age and Fare

# Step 6: Remove Missing Values
df = df.dropna(subset=['Age', 'Fare'])

# Step 7: Group Data by Categorical Variable
grouped_data = df.groupby('Sex')

# Step 8: Calculate Summary Statistics
summary_stats = grouped_data[['Age', 'Fare']].agg(
    ['mean', 'median', 'min', 'max', 'std']
)

# Step 9: Display Summary Statistics
print("\nSummary Statistics Grouped by Sex:")
print(summary_stats)

# Step 10: Create List of Age Values for Each Category
male_age = df[df['Sex'] == 'male']['Age'].tolist()
female_age = df[df['Sex'] == 'female']['Age'].tolist()

# Step 11: Display Lists
print("\nMale Age List:")
print(male_age)

print("\nFemale Age List:")
print(female_age)
```

```python
# Statistical Details of Iris Dataset

# Step 1: Import Libraries
import pandas as pd
import numpy as np

# Step 2: Load Dataset
df = pd.read_csv("iris.csv")

# Step 3: Display First 5 Rows
print("First 5 Rows:")
print(df.head())

# Step 4: Display Dataset Information
print("\nDataset Information:")
print(df.info())

# Step 5: Display Unique Species
print("\nSpecies in Dataset:")
print(df['variety'].unique())

# Step 6: Group Dataset by Species
grouped_data = df.groupby('variety')

# Step 7: Calculate Statistical Details
stats = grouped_data.agg(
    ['mean', 'median', 'std', 'min', 'max']
)

# Step 8: Display Statistical Details
print("\nStatistical Details:")
print(stats)

# Step 9: Calculate Percentiles
for species in df['variety'].unique():

    print(f"\nPercentiles for {species}:")

    species_data = df[df['variety'] == species]

    print(
        species_data[
            ['sepal.length', 'sepal.width',
             'petal.length', 'petal.width']
        ].quantile([0.25, 0.50, 0.75])
    )
```
