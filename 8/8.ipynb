```python
# Data Visualization I
# Titanic Dataset Analysis using Seaborn

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
df = pd.read_csv("titanic.csv")

# Step 3: Display First 5 Rows
print("First 5 Rows of Dataset:")
print(df.head())

# Step 4: Dataset Information
print("\nDataset Information:")
print(df.info())

# Step 5: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# ---------------------------------------------------
# PART 1 : Finding Patterns using Seaborn
# ---------------------------------------------------

# Set Seaborn Style
sns.set_style("whitegrid")

# 1. Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# 2. Survival based on Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival based on Gender")
plt.show()

# 3. Passenger Class Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()

# 4. Survival based on Passenger Class
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival based on Passenger Class")
plt.show()

# 5. Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# ---------------------------------------------------
# PART 2 : Histogram of Fare Column
# ---------------------------------------------------

plt.figure(figsize=(8,5))
plt.hist(df['Fare'], bins=30, edgecolor='black')
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.show()

# ---------------------------------------------------
# Inference / Observations
# ---------------------------------------------------

print("\nInference:")
print("1. Female passengers had a higher survival rate than males.")
print("2. First-class passengers survived more compared to third-class passengers.")
print("3. Most passengers were traveling in third class.")
print("4. Most passengers paid lower fares, while a few paid very high fares.")
print("5. Passenger ages are mostly between 20 and 40 years.")
```
