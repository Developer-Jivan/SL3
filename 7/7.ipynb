```python
# Practical 7: NLP Preprocessing + TF-IDF

import nltk
import pandas as pd
import ssl
from sklearn.feature_extraction.text import TfidfVectorizer

# SSL fix
try:
    _create_unverified_https_context = ssl._create_unverified_context
    ssl._create_default_https_context = _create_unverified_https_context
except:
    pass

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Downloads
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('wordnet', quiet=True)

docs = [
    "Natural language processing helps computers understand human language",
    "Machine learning improves automatically through experience",
    "Data science uses statistics and programming",
    "Artificial intelligence performs smart tasks"
]

text = docs[0].lower()

tokens = word_tokenize(text)
print("Tokens:")
print(tokens)

print("\nPOS Tags:")
print(pos_tag(tokens))

filtered = [w for w in tokens if w not in stopwords.words('english')]
print("\nAfter Stopword Removal:")
print(filtered)

stemmer = PorterStemmer()
print("\nStemming:")
print([stemmer.stem(w) for w in filtered])

lemmatizer = WordNetLemmatizer()
print("\nLemmatization:")
print([lemmatizer.lemmatize(w) for w in filtered])

vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(docs)

tfidf_df = pd.DataFrame(
    tfidf.toarray(),
    columns=vectorizer.get_feature_names_out()
)

print("\nTF-IDF Matrix:")
print(tfidf_df)

idf_df = pd.DataFrame({
    "Word": vectorizer.get_feature_names_out(),
    "IDF Value": vectorizer.idf_
})

print("\nIDF Values:")
print(idf_df)
```
