```python
# Data Visualization II
# Titanic Dataset - Box Plot Analysis

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
df = pd.read_csv("titanic.csv")

# Step 3: Display First 5 Rows
print("First 5 Rows:")
print(df.head())

# Step 4: Dataset Information
print("\nDataset Information:")
print(df.info())

# Step 5: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# ---------------------------------------------------
# Box Plot : Age Distribution with Gender and Survival
# ---------------------------------------------------

# Set style
sns.set_style("whitegrid")

# Create Box Plot
plt.figure(figsize=(10,6))

sns.boxplot(
    x='Sex',
    y='Age',
    hue='Survived',
    data=df
)

# Title and Labels
plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

# Show Plot
plt.show()

# ---------------------------------------------------
# Statistical Analysis
# ---------------------------------------------------

print("\n================ ANALYSIS ================\n")

# Survival count by gender
print("Survival Count by Gender:")
print(df.groupby('Sex')['Survived'].value_counts())
print()

# Age statistics by gender
print("Age Statistics by Gender:")
print(df.groupby('Sex')['Age'].describe())
print()

# Age statistics by survival
print("Age Statistics by Survival:")
print(df.groupby('Survived')['Age'].describe())
print()

# ---------------------------------------------------
# Inference
# ---------------------------------------------------

print("\n================ INFERENCE ================\n")

print("1. Female passengers had a higher survival rate compared to male passengers.")

print("2. Most passengers were adults, mainly between 20 and 40 years of age.")

print("3. Male passengers show a wider spread in age distribution compared to females.")

print("4. Some outliers are visible in the box plot, indicating very young children and elderly passengers.")

print("5. Survivors generally include more females and younger passengers.")
```
