```python
# ============================================
# Practical: Create Database, Table, Insert Data,
# and Run Simple Queries using SQLite (Impala-like)
# ============================================

# Step 1: Import Required Packages
import sqlite3
import pandas as pd

# Step 2: Connect to Database
# ':memory:' creates a temporary database in RAM
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print("Database created successfully!")

# Step 3: Create Table
cursor.execute("""
CREATE TABLE students (
    student_id INTEGER,
    name TEXT,
    age INTEGER,
    course TEXT,
    marks REAL
)
""")

print("Table 'students' created successfully!")

# Step 4: Insert Data into Table
students_data = [
    (1, 'Rohan', 21, 'AI', 85.5),
    (2, 'Amit', 22, 'DS', 78.0),
    (3, 'Priya', 20, 'AI', 92.3),
    (4, 'Sneha', 21, 'ML', 88.8),
    (5, 'Karan', 23, 'DS', 67.5)
]

cursor.executemany(
    "INSERT INTO students VALUES (?, ?, ?, ?, ?)",
    students_data
)

conn.commit()

print("Data inserted successfully!")

# ============================================
# Step 5: Display Full Table
# ============================================

print("\n----- Complete Student Table -----")

df = pd.read_sql_query(
    "SELECT * FROM students",
    conn
)

print(df)

# ============================================
# Step 6: Display Selected Columns
# ============================================

print("\n----- Name and Course -----")

df = pd.read_sql_query(
    "SELECT name, course FROM students",
    conn
)

print(df)

# ============================================
# Step 7: WHERE Clause Example
# ============================================

print("\n----- Students with Marks > 80 -----")

df = pd.read_sql_query(
    "SELECT * FROM students WHERE marks > 80",
    conn
)

print(df)

# ============================================
# Step 8: Aggregate Function Example
# ============================================

print("\n----- Average Marks -----")

df = pd.read_sql_query(
    "SELECT AVG(marks) AS average_marks FROM students",
    conn
)

print(df)

# ============================================
# Step 9: UPDATE Query
# ============================================

cursor.execute("""
UPDATE students
SET marks = 70
WHERE student_id = 5
""")

conn.commit()

print("\n----- Table After UPDATE -----")

df = pd.read_sql_query(
    "SELECT * FROM students",
    conn
)

print(df)

# ============================================
# Step 10: DELETE Query
# ============================================

cursor.execute("""
DELETE FROM students
WHERE student_id = 2
""")

conn.commit()

print("\n----- Table After DELETE -----")

df = pd.read_sql_query(
    "SELECT * FROM students",
    conn
)

print(df)

# ============================================
# Step 11: DROP TABLE
# ============================================

cursor.execute("DROP TABLE students")

print("\nTable dropped successfully!")

# ============================================
# Step 12: Close Connection
# ============================================

conn.close()

print("Connection closed successfully!")
```
