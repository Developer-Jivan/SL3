```python
# Data Visualization III
# Iris Dataset Visualization

# ---------------------------------------------------
# Step 1: Import Libraries
# ---------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------
# Step 2: Load Dataset
# ---------------------------------------------------

df = pd.read_csv("iris.csv")

# ---------------------------------------------------
# Step 3: Display First 5 Rows
# ---------------------------------------------------

print("First 5 Rows:\n")

print(df.head())

# ---------------------------------------------------
# Step 4: Dataset Information
# ---------------------------------------------------

print("\nDataset Information:\n")

print(df.info())

# ---------------------------------------------------
# Step 5: Features and Their Types
# ---------------------------------------------------

print("\nFeatures and Datatypes:\n")

for col in df.columns:

    print(col, ":", df[col].dtype)

# ---------------------------------------------------
# Step 6: Separate Numerical and Categorical Columns
# ---------------------------------------------------

numerical_cols = [
    'sepal.length',
    'sepal.width',
    'petal.length',
    'petal.width'
]

categorical_cols = ['variety']

# ---------------------------------------------------
# Step 7: Histograms for Numerical Features
# ---------------------------------------------------

print("\nCreating Histograms...")

plt.figure(figsize=(12,8))

for i, col in enumerate(numerical_cols, 1):

    plt.subplot(2,2,i)

    plt.hist(
        df[col],
        bins=15,
        edgecolor='black'
    )

    plt.title(f'Histogram of {col}')

    plt.xlabel(col)

    plt.ylabel('Frequency')

plt.tight_layout()

plt.show()

# ---------------------------------------------------
# Step 8: Boxplots for Numerical Features
# ---------------------------------------------------

print("\nCreating Boxplots...")

plt.figure(figsize=(12,8))

for i, col in enumerate(numerical_cols, 1):

    plt.subplot(2,2,i)

    sns.boxplot(y=df[col])

    plt.title(f'Boxplot of {col}')

plt.tight_layout()

plt.show()

# ---------------------------------------------------
# Step 9: Outlier Detection Using IQR Method
# ---------------------------------------------------

print("\nOutlier Detection Using IQR Method")

for col in numerical_cols:

    Q1 = df[col].quantile(0.25)

    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR

    upper_limit = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower_limit) |
        (df[col] > upper_limit)
    ]

    print(f"\nFeature: {col}")

    print("Lower Limit:", lower_limit)

    print("Upper Limit:", upper_limit)

    print("Number of Outliers:", len(outliers))

# ---------------------------------------------------
# Step 10: Inference
# ---------------------------------------------------

print("\nInference:")

print("""
1. All feature columns are continuous numerical variables.

2. The 'variety' column is a nominal categorical variable.

3. Histograms show the distribution of each feature.

4. Boxplots help visualize spread and detect possible outliers.

5. Some features may contain statistical outliers,
   especially sepal.width.

6. These outliers appear to be genuine observations
   rather than data errors.
""")
```
